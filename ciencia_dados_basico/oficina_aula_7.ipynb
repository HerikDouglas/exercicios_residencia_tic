Questão 2: Pesquise as diferenças entre bancos de dados relacionais e NoSQL. Quais tipos de empresas se beneficiam mais de cada um?

- RELACIONAIS
  - Armazenam dados em tabelas compostas por linhas e colunas
  - Cada linha possui uma chave primaria para sua identificação
  - faz relacionamento entre tabelas através de chaves estrangeiras
  - utiliza linguagem SQL
- NÃO RELACIOANIS (NoSQL)
  - Não segue um modelo de tabela linha e coluna
  - Suporta dados não estruturados
  - Ao invés de SQL utiliza-se linguagens como MongoDB
- Empresas que precisam trabalhar com uma grande quantidade de volume de dados não estruturados como redes sociais podem se beneficiar de banco de dados NoSQL. Empresas como as do ramo financeiro que precisam ter seus dados bem organizados, se beneficiam bem mais de banco de dados SQL.
