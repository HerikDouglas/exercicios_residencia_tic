Você é um analista de dados de uma loja on-line e recebeu uma base de dados contendo informações de vendas. Seu objetivo é analisar os dados para entender o perfil dos clientes e o desempenho das vendas. A base de dados contém as seguintes colunas: Nome, Idade, Cidade, Vendas.

Tarefa: Utilize os conhecimentos adquiridos sobre pandas e numpy para o passo a passo a seguir:

1 - Crie um DataFrame com base nos dados fornecidos;

2 - Filtre os clientes que têm mais de 30 anos;

3 - Calcule a média de idade dos clientes;

4 - Substitua valores faltantes na coluna Cidade por ‘Desconhecido’;

5 - Calcule a soma total das vendas.



Dados de Entrada:

dados = {

    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],

    'Idade': [28, 34, 29, None, 42],

    'Cidade': ['São Paulo', 'Rio de Janeiro', None, 'Curitiba', 'Belo Horizonte'],

    'Vendas': [150, 200, 300, 400, None]

}

```python
#Importando bibliotecas
import pandas as pd
import numpy as np
```

1 - Crie um DataFrame com base nos dados fornecidos;

```python
dados = {
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],
    'Idade': [28, 34, 29, None, 42],
    'Cidade': ['São Paulo', 'Rio de Janeiro', None, 'Curitiba', 'Belo Horizonte'],
    'Vendas': [150, 200, 300, 400, None]}

df = pd.DataFrame(dados)
df
```

2 - Filtre os clientes que têm mais de 30 anos

```python
#Verifica o df na coluna Idade quem tem mais de 30 anos criando um novo df e salvando na variável clientes_mais_30
clientes_mais_30 = df[df['Idade'] > 30]
clientes_mais_30
```

3 - Calcule a média de idade dos clientes

```python
#Soma todos os valores da coluna 'Idade' do df e divide pela quantidade de pessoas no df
media_idade = df['Idade'].mean()
print(f"A média de idade dos clientes é {media_idade:.2f} anos")
```

4 - Substitua valores faltantes na coluna Cidade por ‘Desconhecido’

```python
#Verifica falores faltantes na coluna 'Cidade' e os troca por "Desconhecidos" e mostra o df com as devidas mudanças
df['Cidade'].fillna('Desconhecido', inplace=True)
df
```

5 - Calcule a soma total das vendas

```python
#Na coluna 'Vendas' do df soma todos os valares e guarda na variável 'soma_total_vendas'
soma_total_vendas = df['Vendas'].sum()
print(f'A soma total das vendas é R$ {soma_total_vendas:.2f}')
```
